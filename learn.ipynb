CONCATENATION AND MERGE DATAFRAME (JOINS)

```python
df1 = pd.DataFrame({'ID':[1,2,3,4,5], 'Name':['khushbu', 'sneha', 'priya', 'snehal', 'priyanka']})
df1
```

```python
df2 = pd.DataFrame({'ID':[1,2,3,3,5,6], 'Age':[20, 21, 22, 23, 24, 25]})
df2
```

```python
pd.concat([df1,df2],axis=0)
```

```python
pd.concat([df1,df2],axis=1)
```

```python
#merge two dataframes based on common column
pd.merge(df1, df2, on='ID', how='outer') 
```

```python
pd.merge(df1, df2, left_on='ID', right_on='ID', how='outer') 
```
